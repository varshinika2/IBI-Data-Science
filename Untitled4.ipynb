```python
# Import required libraries
import pandas as pd

# Load the dataset (adjust the path if needed)
df = pd.read_csv("online_retail_II_with_marketing.csv")

# Inspect the structure
print(df.shape)
print(df.dtypes)
df.head()
```

```python
#A. Handle Missing Values with Iterative Imputer and Outliers with Z-Score
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from scipy import stats
import numpy as np

# Select numerical columns
num_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Impute missing values
imputer = IterativeImputer(random_state=42)
df[num_cols] = imputer.fit_transform(df[num_cols])

# Remove outliers using Z-score method
z_scores = np.abs(stats.zscore(df[["Quantity", "Price"]]))
df = df[(z_scores < 3).all(axis=1)].reset_index(drop=True)
```

```python
#B. Normalize with StandardScaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
#C. Feature Engineering
# Return Rate = returns / total per product
returns = df[df["Quantity"] < 0].groupby("StockCode")["Quantity"].count()
total_orders = df[df["Quantity"] > 0].groupby("StockCode")["Quantity"].count()
return_rate = (returns / total_orders).fillna(0)
df["Return Rate"] = df["StockCode"].map(return_rate)

# Product Profitability = Price * Quantity
df["Product Profitability"] = df["Price"] * df["Quantity"]

# Customer Engagement Score = # of invoices per customer (normalized)
df["Engagement Score"] = df.groupby("Customer ID")["Invoice"].transform("count")
df["Engagement Score"] = scaler.fit_transform(df[["Engagement Score"]])

# Marketing Campaign Effectiveness
df["Campaign Effectiveness"] = df["Conversion Rate"] * df["Customer Reach"] / 100
```

```python
# RFM Feature Engineering (Recency, Frequency, Monetary)
import numpy as np
from datetime import datetime

# Step 1: Clean and filter valid transactions
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"], errors='coerce')
df = df[df["Quantity"] > 0]
df = df[df["Customer ID"].notnull()]

# Step 2: Create RFM Table
snapshot_date = df["InvoiceDate"].max() + pd.Timedelta(days=1)
rfm = df.groupby("Customer ID").agg({
    "InvoiceDate": lambda x: (snapshot_date - x.max()).days,
    "Invoice": "nunique",
    "Price": lambda x: (x * df.loc[x.index, "Quantity"]).sum()
}).reset_index()

rfm.columns = ["CustomerID", "Recency", "Frequency", "Monetary"]
rfm.head()
```

```python
#Clustering (e.g. KMeans)
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Step 3: Normalize
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm[["Recency", "Frequency", "Monetary"]])

# Step 4: Optimal K using Elbow Method
wcss = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(rfm_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 10), wcss, marker="o")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method for Optimal K")
plt.show()
```

```python
#Assign Clusters
# Step 5: Fit with optimal K (e.g., K=4)
kmeans = KMeans(n_clusters=4, random_state=42)
rfm["Cluster"] = kmeans.fit_predict(rfm_scaled)

rfm.groupby("Cluster").agg({
    "Recency": "mean",
    "Frequency": "mean",
    "Monetary": "mean",
    "CustomerID": "count"
}).rename(columns={"CustomerID": "Num_Customers"})
```

```python
#Dynamic Pricing Strategy
# Step 1: Read only necessary columns in chunks
import pandas as pd

# 1. Upload file if using Google Colab (uncomment this if needed)
# from google.colab import files
# uploaded = files.upload()

# 2. Set filename  replace with correct path or filename after upload
filename = "online_retail_II_with_marketing.csv"

# 3. Only read relevant columns
use_cols = ["StockCode", "Description", "Invoice", "InvoiceDate", "Quantity", "Price", "Customer ID"]

# 4. Read in chunks and filter
chunks = []

for chunk in pd.read_csv(filename, usecols=use_cols, chunksize=100000):
    # Keep only valid rows: positive quantity, positive price, and valid customer ID
    filtered = chunk[
        (chunk["Quantity"] > 0) &
        (chunk["Price"] > 0) &
        (chunk["Customer ID"].notna())
    ]
    chunks.append(filtered)

# 5. Concatenate chunks into a single DataFrame
df = pd.concat(chunks, ignore_index=True)

# 6. Convert date column
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"], errors="coerce")

# Preview
df.head()

```

```python
product_demand = df.groupby(["StockCode", "Description", "Price"]).agg({
    "Quantity": "sum",
    "Invoice": "nunique"
}).reset_index()

product_demand.rename(columns={"Invoice": "Transactions"}, inplace=True)
```

```python
price_point_counts = product_demand.groupby("StockCode")["Price"].count().reset_index()
multi_price_products = price_point_counts[price_point_counts["Price"] >= 5]["StockCode"]

elastic_df = product_demand[product_demand["StockCode"].isin(multi_price_products)]
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

elasticity_results = []

for stockcode in elastic_df["StockCode"].unique():
    temp = elastic_df[elastic_df["StockCode"] == stockcode]
    if temp["Quantity"].gt(0).all() and temp["Price"].gt(0).all():
        X_log = np.log(temp[["Price"]])
        y_log = np.log(temp["Quantity"])

        model = LinearRegression()
        model.fit(X_log, y_log)

        elasticity = model.coef_[0]  # Demand elasticity
        r2 = model.score(X_log, y_log)

        elasticity_results.append({
            "StockCode": stockcode,
            "Description": temp["Description"].iloc[0],
            "Elasticity": elasticity,
            "R2": r2,
            "NumPoints": len(temp)
        })

elasticity_df = pd.DataFrame(elasticity_results)
```

```python
def classify_elasticity(e):
    if e <= -1:
        return "Elastic"
    elif -1 < e < 0:
        return "Inelastic"
    else:
        return "Abnormal"

elasticity_df["ElasticityType"] = elasticity_df["Elasticity"].apply(classify_elasticity)
elasticity_df.sort_values(by="Elasticity").head(10)
```

```python
import pandas as pd
import plotly.express as px

# Load the dataset
df = pd.read_csv("online_retail_II_with_marketing.csv")

#  Convert InvoiceDate to datetime explicitly
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')

# Clean and sample
df = df[['InvoiceDate', 'Customer ID', 'Quantity', 'Price']].dropna()
df = df[(df['Quantity'] > 0) & (df['Price'] > 0)].copy()
df['TotalSales'] = df['Quantity'] * df['Price']
df = df.sample(n=50000, random_state=1)

#  Recompute latest date
latest_date = df['InvoiceDate'].max()

# Compute recency for each customer
recency_series = df.groupby('Customer ID')['InvoiceDate'].max()
recency_days = (latest_date - recency_series).dt.days

# Compute frequency and monetary
frequency = df.groupby('Customer ID')['InvoiceDate'].count()
monetary = df.groupby('Customer ID')['TotalSales'].sum()

# Combine into RFM
rfm = pd.DataFrame({
    'Recency': recency_days,
    'Frequency': frequency,
    'Monetary': monetary
}).dropna()

# RFM Scoring
rfm['R_Score'] = pd.qcut(rfm['Recency'], 4, labels=[4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 4, labels=[1,2,3,4])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 4, labels=[1,2,3,4])

rfm['RFM_Segment'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
rfm['Segment'] = 'Low-Value'
rfm.loc[rfm['RFM_Segment'] >= '343', 'Segment'] = 'Mid-Value'
rfm.loc[rfm['RFM_Segment'] >= '444', 'Segment'] = 'High-Value'

# Reset for plotting
rfm.reset_index(inplace=True)

# Plot
fig = px.scatter(rfm, x='Recency', y='Monetary', color='Segment', size='Frequency',
                 title='Customer Segmentation based on RFM')
fig.show()
```
